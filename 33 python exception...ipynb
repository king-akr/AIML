```python
lst = [2, 0 , "hello" , None]
for ele in lst:
    try:
        ele = int(ele)
        print(6/ele)
    except Exception as e:
        print(e)
        
    print("-"*80)
    
print("100 lines of code")
```

```python
lst = [2, 0 , "hello" , None]
for ele in lst:
    try:
        ele = int(ele)
        print(6/ele)
    except Exception as e:
        print(e)
        print(sys.exc_info()[0])
        
    print("-"*80)
    
print("100 lines of code")
```
