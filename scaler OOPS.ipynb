```python
class Account:
    acc_no = 1001

    def __init__(self, balance):
        print("Creating new account")
        self.accno = Account.acc_no
        Account.acc_no += 1
        self.acc_balance = balance
        self.max_trans = 0
        self.num_trans = 0

    def __repr__(self):
        return f"Acc{self.accno} has balance {self.acc_balance}"

    def deposit(self, amount):
        if amount > 0 and self.num_trans < self.max_trans:
            print("Adding money....")
            self.acc_balance += amount
            self.num_trans += 1
        else:
            print("Deposit limit crossed or invalid amount")

    def withdrawl(self, amount):
        if self.num_trans >= self.max_trans:
            print("Transaction limit crossed")
        elif amount > self.acc_balance:
            print("Insufficient balance")
        else:
            print("Withdrawing money....")
            self.acc_balance -= amount
            self.num_trans += 1

    def getInterest(self):
        pass


class SavingAccount(Account):
    def __init__(self, balance):
        super().__init__(balance)
        self.max_trans = 2

    def getInterest(self):
        interest = 0.05 * self.acc_balance
        print(f"Interest earned: {interest}")


class CurrentAccount(Account):
    def __init__(self, balance):
        super().__init__(balance)
        self.max_trans = 10

    def getInterest(self):
        interest = 0.03 * self.acc_balance
        print(f"Interest earned: {interest}")


# # -------------------- TESTING --------------------

# sa_1 = SavingAccount(5000)
# print(sa_1)

# sa_1.deposit(1000)
# sa_1.deposit(500)
# sa_1.deposit(200)   # limit crossed

# sa_1.withdrawl(1000)

# sa_1.getInterest()
# print(sa_1)

# print("---------------")

# ca_1 = CurrentAccount(10000)
# ca_1.deposit(2000)
# ca_1.withdrawl(3000)
# ca_1.getInterest()
# print(ca_1)
```
